```python
!pip install pandas numpy opencv-python matplotlib cellpose scikit-image scikit-learn seaborn
# 1. IMPORTS
import pandas as pd
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
from cellpose import models
from skimage.measure import regionprops, label
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.manifold import TSNE
import seaborn as sns
sns.set(style='whitegrid')

# 2. SIMULATE METADATA
# Simulate 10 image names and drug labels
metadata = pd.DataFrame({
    'Image_FileName_DNA': [f'image_{i}.png' for i in range(10)],
    'Metadata_pert_iname': np.random.choice(['DrugA', 'DrugB', 'DrugC'], size=10)
})

# 3. GENERATE RANDOM GRAYSCALE IMAGES AND SAVE
image_dir = 'test_images'
os.makedirs(image_dir, exist_ok=True)

for fname in metadata['Image_FileName_DNA']:
    img = np.random.randint(0, 255, size=(128, 128), dtype=np.uint8)
    cv2.imwrite(os.path.join(image_dir, fname), img)

# 4. LOAD IMAGES
images = []
drug_labels = []

for _, row in metadata.iterrows():
    filename = row['Image_FileName_DNA']
    drug_label = row['Metadata_pert_iname']
    img_path = os.path.join(image_dir, filename)
    if os.path.exists(img_path):
        img = load_image(img_path)
        images.append(img)
        drug_labels.append(drug_label)

# 5. SEGMENT USING CELLPOSE
model = models.Cellpose(model_type='cyto')

masks_list = []
for img in images:
    masks, flows, styles, diams = model.eval(img, diameter=None, channels=[0, 0])
    masks_list.append(masks)

# 6. EXTRACT FEATURES
from skimage.measure import regionprops, label as sklabel  # rename to avoid conflict

def extract_features(mask, img):
    props = regionprops(sklabel(mask), intensity_image=img)
    features = []
    for p in props:
        features.append([
            p.area,
            p.perimeter,
            p.mean_intensity,
            p.eccentricity,
            p.solidity
        ])
    if len(features) == 0:
        return np.zeros(5)
    return np.mean(features, axis=0)

X = np.array([extract_features(m, i) for m, i in zip(masks_list, images)])
y = np.array(drug_labels)

# 7. CLASSIFIER
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))

plt.figure(figsize=(8, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d',
            xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# 8. T-SNE VISUALIZATION
tsne = TSNE(n_components=2, perplexity=3, random_state=42)
X_embedded = tsne.fit_transform(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_embedded[:, 0], y=X_embedded[:, 1], hue=y, palette='tab10')
plt.title('t-SNE of Cell Features')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
